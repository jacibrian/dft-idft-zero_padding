```python
### Lab N.7 Realizado por Jorge Cibrián

# Librerias
import numpy as np
import matplotlib.pyplot as plt

# Funciones
def dft(x, N):
    n = np.arange(N)  # vector de tiempo
    k = np.arange(N)  # vector de frecuencia
    WN = np.exp(-1j * 2 * np.pi / N)
    nk = np.outer(n, k)
    WNnk = WN ** nk
    Xk = np.dot(x, WNnk)
    return Xk, k
```

```python
### EJERCICIO 2

# Frecuencias
Fm=8000;f1=1000;f2=400;Tm=1/Fm;N=128;
# Valores discretos donde evaluaremos la función:
t=np.linspace(0,N*Tm,N,endpoint=False,retstep=True)
# Verificando que el tamano de paso de t es Tm
# con la opcion retstep
t[1]==Tm
```

```python
# Ahora asignando t como el arreglo de valores de t
t=t[0]
# Evaluando
x=np.cos(2*np.pi*f1*t)+np.cos(2*np.pi*f2*t)
```

```python
# Calculando la trans. discreta de Fourier
X=dft(x,N)
Xf=X[0]

# Magnitud de la trans.:
Xf_mag=np.absolute(Xf)
# Fase de la trans.:
Xf_fas=np.angle(Xf)
# 128 muestras con Fm=8000 Hz:
f=np.linspace(0,Fm,N,endpoint=False)
```

```python
fig, ejes = plt.subplots(3, 1, figsize=(8, 7))
(p1,p2,p3) = ejes.flatten()

p1.plot(t,x);
p1.axvline(0, color='black', linewidth=0.8, linestyle='--');
p1.grid(True, linestyle='--', alpha=0.7)
p1.set_title('x=cos(2pi f1 t)+cos(2pi f2 t)');
p1.set_xlim();p1.set_xlabel('tiempo (s)');
plt.subplots_adjust(hspace=0.6);

p2.stem(f,Xf_mag,basefmt=" ");
p2.axvline(0, color='black', linewidth=0.8, linestyle='--');
p2.grid(True, linestyle='--', alpha=0.7)
p2.set_title('Magnitud de X(f) con N=128 muestras');
p2.set_xlim();p1.set_xlabel('frec. (Hz)');
plt.subplots_adjust(hspace=0.6);

p3.stem(f,Xf_fas,basefmt=" ");
p3.axvline(0, color='black', linewidth=0.8, linestyle='--');
p3.grid(True, linestyle='--', alpha=0.7)
p3.set_title('Fase de X(f) con N=128 muestras');
p3.set_xlim();p1.set_xlabel('frec. (Hz)');
plt.subplots_adjust(hspace=0.6);

plt.show()
```

```python
### EJERCICIO 7

def idft(X):
    N = len(X)
    n = np.arange(N)
    k = np.arange(N)
    WN = np.exp(-1j * 2 * np.pi / N)
    nk = np.outer(n, k)
    WNnk = WN ** (-nk)
    x = np.dot(X, WNnk) / N
    return x, n
```
